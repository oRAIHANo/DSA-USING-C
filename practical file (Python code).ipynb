# 1) Implement linear search and determine the time required to search an element. Repeat the experiment for different values of n, the number of elements in the list to be searched and plot a graph of the time taken versus n.

```python
import time
import matplotlib.pyplot as plt

# Function to perform linear search
def linear_search(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

# Function to perform the experiment and record the time taken for different n values
def experiment():
    n_values = []
    time_values = []
    
    # Range of n values for the experiment
    for n in range(1000, 100001, 5000):
        arr = list(range(1, n+1))  # Create a list from 1 to n
        x = -1  # Element to search for (not in the list to simulate worst-case scenario)
        
        # Record the start time
        start_time = time.time()
        
        # Perform linear search
        linear_search(arr, x)
        
        # Record the end time
        end_time = time.time()
        
        # Calculate time taken
        time_taken = end_time - start_time
        
        # Store results
        n_values.append(n)
        time_values.append(time_taken)
        print(f"n = {n}, Time taken = {time_taken:.6f} seconds")
    
    return n_values, time_values

# Function to plot the graph
def plot_graph(n_values, time_values):
    plt.plot(n_values, time_values, marker='o')
    plt.title("Linear Search Time Complexity")
    plt.xlabel("Number of elements (n)")
    plt.ylabel("Time taken (seconds)")
    plt.grid(True)
    plt.show()

# Run the experiment and plot the results
n_values, time_values = experiment()
plot_graph(n_values, time_values)
```

# 2) Implement recursive binary search and determine the time required to search an element. Repeat the experiment for different values of n, the number of elements in the list to be searched and plot a graph of the time taken versus n.

```python
import time
import matplotlib.pyplot as plt

# Recursive Binary Search function
def recursive_binary_search(arr, left, right, x):
    if right >= left:
        mid = left + (right - left) // 2
        
        # If the element is present at the middle
        if arr[mid] == x:
            return mid
        
        # If the element is smaller than mid, it can only be in the left subarray
        elif arr[mid] > x:
            return recursive_binary_search(arr, left, mid - 1, x)
        
        # Otherwise, it can only be in the right subarray
        else:
            return recursive_binary_search(arr, mid + 1, right, x)
    
    # If the element is not present
    return -1

# Function to perform the experiment and record the time taken for different n values
def experiment():
    n_values = []
    time_values = []
    
    # Range of n values for the experiment
    for n in range(1000, 100001, 5000):
        arr = list(range(1, n+1))  # Create a sorted list from 1 to n
        x = -1  # Element to search for (not in the list to simulate worst-case scenario)
        
        # Record the start time
        start_time = time.time()
        
        # Perform recursive binary search
        recursive_binary_search(arr, 0, len(arr) - 1, x)
        
        # Record the end time
        end_time = time.time()
        
        # Calculate time taken
        time_taken = end_time - start_time
        
        # Store results
        n_values.append(n)
        time_values.append(time_taken)
        print(f"n = {n}, Time taken = {time_taken:.6f} seconds")
    
    return n_values, time_values

# Function to plot the graph
def plot_graph(n_values, time_values):
    plt.plot(n_values, time_values, marker='o')
    plt.title("Recursive Binary Search Time Complexity")
    plt.xlabel("Number of elements (n)")
    plt.ylabel("Time taken (seconds)")
    plt.grid(True)
    plt.show()

# Run the experiment and plot the results
n_values, time_values = experiment()
plot_graph(n_values, time_values)
```

# 3) Sort a given set of elements using heap sort method and determine the time required to sort the elements. Repeat the experiment for different values of n, the number of elements in the list to be sorted and plot a graph of the time taken versus n.

```python
import time
import matplotlib.pyplot as plt

# Function to heapify a subtree rooted with node i
def heapify(arr, n, i):
    largest = i  # Initialize largest as root
    left = 2 * i + 1  # Left child
    right = 2 * i + 2  # Right child

    # If left child is larger than root
    if left < n and arr[left] > arr[largest]:
        largest = left

    # If right child is larger than largest so far
    if right < n and arr[right] > arr[largest]:
        largest = right

    # If largest is not root
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # Swap

        # Recursively heapify the affected subtree
        heapify(arr, n, largest)

# Heap sort function
def heap_sort(arr):
    n = len(arr)

    # Build a max heap
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extract elements from the heap
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # Swap
        heapify(arr, i, 0)

# Function to perform the experiment and record the time taken for different n values
def experiment():
    n_values = []
    time_values = []
    
    # Range of n values for the experiment
    for n in range(1000, 100001, 5000):
        arr = list(range(n, 0, -1))  # Create a list in descending order (worst case for heap sort)
        
        # Record the start time
        start_time = time.time()
        
        # Perform heap sort
        heap_sort(arr)
        
        # Record the end time
        end_time = time.time()
        
        # Calculate time taken
        time_taken = end_time - start_time
        
        # Store results
        n_values.append(n)
        time_values.append(time_taken)
        print(f"n = {n}, Time taken = {time_taken:.6f} seconds")
    
    return n_values, time_values

# Function to plot the graph
def plot_graph(n_values, time_values):
    plt.plot(n_values, time_values, marker='o')
    plt.title("Heap Sort Time Complexity")
    plt.xlabel("Number of elements (n)")
    plt.ylabel("Time taken (seconds)")
    plt.grid(True)
    plt.show()

# Run the experiment and plot the results
n_values, time_values = experiment()
plot_graph(n_values, time_values)
```

# 4) Sort a given set of elements using merge sort method and determine the time required to sort the elements. Repeat the experiment for different values of n, the number of elements in the list to be sorted and plot a graph of the time taken versus n.

```python
import time
import matplotlib.pyplot as plt

# Merge function to merge two halves
def merge(array, low, mid, high):
    # Create temporary merged array
    merged_array = []
    i = low
    j = mid + 1

    # Merge the two halves
    while i <= mid and j <= high:
        if array[i] < array[j]:
            merged_array.append(array[i])
            i += 1
        else:
            merged_array.append(array[j])
            j += 1

    # Copy remaining elements of the left half
    while i <= mid:
        merged_array.append(array[i])
        i += 1

    # Copy remaining elements of the right half
    while j <= high:
        merged_array.append(array[j])
        j += 1

    # Copy merged elements back into the original array
    for k, val in enumerate(merged_array):
        array[low + k] = val

def merge_sort(array, low, high):
    if low < high:
        mid = (low + high) // 2
        merge_sort(array, low, mid)  # Sort the first half
        merge_sort(array, mid + 1, high)  # Sort the second half
        merge(array, low, mid, high)  # Merge the sorted halves

# Function to perform the experiment and record the time taken for different n values
def experiment():
    n_values = []
    time_values = []
    
    # Range of n values for the experiment
    for n in range(1000, 100001, 5000):
        arr = list(range(n, 0, -1))  # Create a list in descending order (worst case for merge sort)
        
        # Record the start time
        start_time = time.time()
        
        # Perform merge sort
        merge_sort(arr, 0, len(arr) - 1)
        
        # Record the end time
        end_time = time.time()
        
        # Calculate time taken
        time_taken = end_time - start_time
        
        # Store results
        n_values.append(n)
        time_values.append(time_taken)
        print(f"n = {n}, Time taken = {time_taken:.6f} seconds")
    
    return n_values, time_values

# Function to plot the graph
def plot_graph(n_values, time_values):
    plt.plot(n_values, time_values, marker='o')
    plt.title("Merge Sort Time Complexity")
    plt.xlabel("Number of elements (n)")
    plt.ylabel("Time taken (seconds)")
    plt.grid(True)
    plt.show()

# Run the experiment and plot the results
n_values, time_values = experiment()
plot_graph(n_values, time_values)
```

# 6) Sort a given set of elements using insertion sort method and determine the time required to sort the elements. Repeat the experiment for different values of n, the number of elements in the list to be sorted and plot a graph of the time taken versus n.

```python
import time
import matplotlib.pyplot as plt

# Function to perform insertion sort
def insertion_sort(array):
    size = len(array)
    for i in range(1, size):
        key = array[i]
        j = i - 1
        
        # Shift elements of array that are greater than the key
        while j >= 0 and array[j] > key:
            array[j + 1] = array[j]
            j -= 1
        array[j + 1] = key

# Function to perform the experiment and record the time taken for different n values
def experiment():
    n_values = []
    time_values = []
    
    # Range of n values for the experiment
    for n in range(1000, 10001, 1000):
        arr = list(range(n, 0, -1))  # Create a list in descending order (worst case for insertion sort)
        
        # Record the start time
        start_time = time.time()
        
        # Perform insertion sort
        insertion_sort(arr)
        
        # Record the end time
        end_time = time.time()
        
        # Calculate time taken
        time_taken = end_time - start_time
        
        # Store results
        n_values.append(n)
        time_values.append(time_taken)
        print(f"n = {n}, Time taken = {time_taken:.6f} seconds")
    
    return n_values, time_values

# Function to plot the graph
def plot_graph(n_values, time_values):
    plt.plot(n_values, time_values, marker='o')
    plt.title("Insertion Sort Time Complexity")
    plt.xlabel("Number of elements (n)")
    plt.ylabel("Time taken (seconds)")
    plt.grid(True)
    plt.show()

# Run the experiment and plot the results
n_values, time_values = experiment()
plot_graph(n_values, time_values)
```

# 7) Sort a given set of elements using Quick sort method and determine the time required to sort the elements. Repeat the experiment for different values of n, the number of elements in the list to be sorted and plot a graph of the time taken versus n.

```python
import time
import matplotlib.pyplot as plt

# Quick sort function
def quick_sort(array, low, high):
    if low < high:
        # pi is the partitioning index, array[pi] is now in the right place
        pi = partition(array, low, high)

        # Recursively sort elements before and after partition
        quick_sort(array, low, pi - 1)
        quick_sort(array, pi + 1, high)

# Partition function for Quick Sort
def partition(array, low, high):
    pivot = array[low]  # Set the first element as pivot
    i = low           # Start from the low index
    j = high + 1      # Start from the high index

    while True:
        # Move i to the right as long as elements are less than or equal to pivot
        i += 1
        while i <= high and array[i] <= pivot:
            i += 1

        # Move j to the left as long as elements are greater than pivot
        j -= 1
        while array[j] > pivot:
            j -= 1

        # If the pointers cross, break out of the loop
        if i >= j:
            break

        # Swap array[i] and array[j]
        array[i], array[j] = array[j], array[i]

    # Swap pivot with array[j] to place the pivot in the correct position
    array[low], array[j] = array[j], array[low]

    return j  # Return the partition index

# Function to perform the experiment and record the time taken for different n values
def experiment():
    n_values = []
    time_values = []
    
    # Range of n values for the experiment
    for n in range(1000, 10001, 1000):
        arr = list(range(n, 0, -1))  # Create a list in descending order (worst case for quicksort)
        
        # Record the start time
        start_time = time.time()
        
        # Perform quick sort
        quick_sort(arr, 0, len(arr) - 1)
        
        # Record the end time
        end_time = time.time()
        
        # Calculate time taken
        time_taken = end_time - start_time
        
        # Store results
        n_values.append(n)
        time_values.append(time_taken)
        print(f"n = {n}, Time taken = {time_taken:.6f} seconds")
    
    return n_values, time_values

# Function to plot the graph
def plot_graph(n_values, time_values):
    plt.plot(n_values, time_values, marker='o')
    plt.title("Quick Sort Time Complexity")
    plt.xlabel("Number of elements (n)")
    plt.ylabel("Time taken (seconds)")
    plt.grid(True)
    plt.show()

# Run the experiment and plot the results
n_values, time_values = experiment()
plot_graph(n_values, time_values)
```

# 8) Sort a given set of elements using Radix sort method and determine the time required to sort the elements. Repeat the experiment for different values of n, the number of elements in the list to be sorted and plot a graph of the time taken versus n.

```python
import time
import matplotlib.pyplot as plt

# A utility function to get the maximum value in array
def get_max(arr):
    return max(arr)

# A function to do counting sort of arr[] according to the digit represented by exp
def counting_sort(arr, exp):
    n = len(arr)
    
    # The output array that will have sorted arr
    output = [0] * n
    count = [0] * 10  # Initialize count array to store count of digits
    
    # Store count of occurrences in count[]
    for i in range(n):
        index = (arr[i] // exp) % 10
        count[index] += 1

    # Change count[i] so that it now contains the actual position of the digit in output[]
    for i in range(1, 10):
        count[i] += count[i - 1]

    # Build the output array
    for i in range(n - 1, -1, -1):
        index = (arr[i] // exp) % 10
        output[count[index] - 1] = arr[i]
        count[index] -= 1

    # Copy the output array to arr[], so that arr now contains sorted numbers according to the current digit
    for i in range(n):
        arr[i] = output[i]

# The main function to sort array using Radix Sort
def radix_sort(arr):
    # Find the maximum number to know the number of digits
    max_val = get_max(arr)

    # Do counting sort for every digit, starting from least significant digit to most significant digit
    exp = 1
    while max_val // exp > 0:
        counting_sort(arr, exp)
        exp *= 10

# Function to perform the experiment and record the time taken for different n values
def experiment():
    n_values = []
    time_values = []
    
    # Range of n values for the experiment
    for n in range(1000, 10001, 1000):
        arr = list(range(n, 0, -1))  # Create a list in descending order (worst case for some algorithms)
        
        # Record the start time
        start_time = time.time()
        
        # Perform radix sort
        radix_sort(arr)
        
        # Record the end time
        end_time = time.time()
        
        # Calculate time taken
        time_taken = end_time - start_time
        
        # Store results
        n_values.append(n)
        time_values.append(time_taken)
        print(f"n = {n}, Time taken = {time_taken:.6f} seconds")
    
    return n_values, time_values

# Function to plot the graph
def plot_graph(n_values, time_values):
    plt.plot(n_values, time_values, marker='o')
    plt.title("Radix Sort Time Complexity")
    plt.xlabel("Number of elements (n)")
    plt.ylabel("Time taken (seconds)")
    plt.grid(True)
    plt.show()

# Run the experiment and plot the results
n_values, time_values = experiment()
plot_graph(n_values, time_values)
```

# 9) Find the minimum cost spanning tree of a given undirected graph using Kruskal's algorithm.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Class to represent a graph
class Graph:
    def __init__(self, vertices):
        self.V = vertices  # Number of vertices
        self.graph = []  # List to store all edges in the format (weight, u, v)

    # Add edge to the graph
    def add_edge(self, u, v, weight):
        self.graph.append((weight, u, v))

    # Utility function to find set (parent) of an element i (uses path compression)
    def find(self, parent, i):
        if parent[i] == i:
            return i
        else:
            parent[i] = self.find(parent, parent[i])
            return parent[i]

    # Utility function to do union of two sets (uses union by rank)
    def union(self, parent, rank, x, y):
        xroot = self.find(parent, x)
        yroot = self.find(parent, y)

        if rank[xroot] < rank[yroot]:
            parent[xroot] = yroot
        elif rank[xroot] > rank[yroot]:
            parent[yroot] = xroot
        else:
            parent[yroot] = xroot
            rank[xroot] += 1

    # Function to construct MST using Kruskal's algorithm
    def kruskal_mst(self):
        result = []  # This will store the resultant MST

        # Step 1: Sort all edges in non-decreasing order of their weight
        self.graph = sorted(self.graph, key=lambda item: item[0])

        parent = []
        rank = []

        # Step 2: Create V subsets with single elements
        for node in range(self.V):
            parent.append(node)
            rank.append(0)

        # Number of edges in MST will be equal to V-1
        e = 0  # Index variable, used for result[]
        i = 0  # Index variable, used to pick the next edge

        # Step 3: Pick the smallest edge and add it to the MST, ensuring no cycle is formed
        while e < self.V - 1:
            weight, u, v = self.graph[i]
            i += 1
            x = self.find(parent, u)
            y = self.find(parent, v)

            # If including this edge doesn't cause a cycle, include it in the result
            if x != y:
                result.append((u, v, weight))
                e += 1
                self.union(parent, rank, x, y)

        # Print the result
        print("Edges in the Minimum Spanning Tree:")
        for u, v, weight in result:
            print(f"{u} -- {v} == {weight}")

        # Return the minimum cost of the spanning tree
        min_cost = sum([weight for _, _, weight in result])
        print(f"Minimum cost of spanning tree: {min_cost}")

        # Visualization of the graph and the MST
        self.visualize_graph(result)

        return min_cost

    def visualize_graph(self, mst_edges):
        G = nx.Graph()
        
        # Add edges to the graph
        for weight, u, v in self.graph:
            G.add_edge(u, v, weight=weight)
        
        # Create a layout for the nodes
        pos = nx.spring_layout(G)

        # Draw the original graph
        nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold')
        
        # Draw the edges of the MST
        mst_edge_set = {(u, v) for u, v, _ in mst_edges}
        nx.draw_networkx_edges(G, pos, edgelist=mst_edge_set, edge_color='red', width=2)
        
        # Add edge labels
        edge_labels = nx.get_edge_attributes(G, 'weight')
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

        plt.title("Graph with Minimum Spanning Tree (MST) highlighted")
        plt.show()

# Example usage:
g = Graph(4)
g.add_edge(0, 1, 10)
g.add_edge(0, 2, 6)
g.add_edge(0, 3, 5)
g.add_edge(1, 3, 15)
g.add_edge(2, 3, 4)

# Function call
g.kruskal_mst()
```

# 10) Find the minimum cost spanning tree of a given undirected graph using Prim's algorithm.

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

class Graph:
    def __init__(self, vertices):
        self.V = vertices  # Number of vertices
        self.graph = np.zeros((vertices, vertices))  # Adjacency matrix

    # Add edge to the graph
    def add_edge(self, u, v, weight):
        self.graph[u][v] = weight
        self.graph[v][u] = weight  # Because the graph is undirected

    # Function to construct MST using Prim's algorithm
    def prim_mst(self):
        # Array to track vertices included in MST
        selected = [False] * self.V
        # Array to store the minimum edge weights
        min_edge = [float('inf')] * self.V
        # Array to store the parent vertices
        parent = [-1] * self.V

        # Start from the first vertex
        min_edge[0] = 0  # Include the first vertex in MST
        parent[0] = -1  # First node is always the root of the MST

        for _ in range(self.V):
            # Find the vertex with the minimum edge weight
            min_weight = float('inf')
            u = -1
            for v in range(self.V):
                if min_edge[v] < min_weight and not selected[v]:
                    min_weight = min_edge[v]
                    u = v
            
            # Add the selected vertex to the MST
            selected[u] = True
            
            # Update the min_edge and parent arrays
            for v in range(self.V):
                if self.graph[u][v] > 0 and not selected[v] and self.graph[u][v] < min_edge[v]:
                    min_edge[v] = self.graph[u][v]
                    parent[v] = u

        # Print the result
        result = []
        for i in range(1, self.V):
            result.append((parent[i], i, self.graph[parent[i]][i]))
            print(f"{parent[i]} -- {i} == {self.graph[parent[i]][i]}")

        return result

    def visualize_graph(self, mst_edges):
        G = nx.Graph()
        # Add edges to the graph
        for i in range(self.V):
            for j in range(i + 1, self.V):
                if self.graph[i][j] != 0:
                    G.add_edge(i, j, weight=self.graph[i][j])

        # Draw the original graph
        pos = nx.spring_layout(G)  # positions for all nodes
        edges = G.edges(data=True)

        # Draw the original graph
        nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800)
        labels = nx.get_edge_attributes(G, 'weight')
        nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

        # Highlight the MST
        mst_edges = [(u, v) for u, v, _ in mst_edges]
        nx.draw_networkx_edges(G, pos, edgelist=mst_edges, edge_color='red', width=2)

        plt.title("Minimum Spanning Tree using Prim's Algorithm")
        plt.show()


# Example usage
g = Graph(5)
g.add_edge(0, 1, 2)
g.add_edge(0, 3, 6)
g.add_edge(1, 2, 3)
g.add_edge(1, 3, 8)
g.add_edge(1, 4, 5)
g.add_edge(2, 4, 7)
g.add_edge(3, 4, 9)

# Function call
mst_edges = g.prim_mst()
g.visualize_graph(mst_edges)
```

# 11) From a given vertex in a weighted connected graph, find shortest paths to other vertices using Dijkstra's algorithm.

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import heapq

class Graph:
    def __init__(self, vertices):
        self.V = vertices  # Number of vertices
        self.graph = nx.Graph()  # Create a Graph object

    # Add edge to the graph
    def add_edge(self, u, v, weight):
        self.graph.add_edge(u, v, weight=weight)

    # Function to find shortest paths using Dijkstra's algorithm
    def dijkstra(self, start):
        # Initialize distances and priority queue
        distances = {vertex: float('infinity') for vertex in self.graph.nodes}
        distances[start] = 0
        priority_queue = [(0, start)]  # (distance, vertex)

        while priority_queue:
            current_distance, current_vertex = heapq.heappop(priority_queue)

            # Nodes can only get added once to the priority queue
            if current_distance > distances[current_vertex]:
                continue

            # Explore neighbors
            for neighbor, attrs in self.graph[current_vertex].items():
                weight = attrs['weight']
                distance = current_distance + weight

                # Only consider this new path if it's better
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    heapq.heappush(priority_queue, (distance, neighbor))

        return distances

    def visualize_graph(self, start, distances):
        pos = nx.spring_layout(self.graph)  # positions for all nodes

        # Draw the original graph
        nx.draw(self.graph, pos, with_labels=True, node_color='lightblue', node_size=800)
        labels = nx.get_edge_attributes(self.graph, 'weight')
        nx.draw_networkx_edge_labels(self.graph, pos, edge_labels=labels)

        # Highlight the shortest paths
        for vertex in distances:
            if distances[vertex] < float('infinity'):
                nx.draw_networkx_nodes(self.graph, pos, nodelist=[vertex], node_color='red')

        plt.title(f"Shortest Paths from Vertex {start} using Dijkstra's Algorithm")
        plt.show()


# Example usage
g = Graph(6)
g.add_edge(0, 1, 7)
g.add_edge(0, 2, 9)
g.add_edge(0, 5, 14)
g.add_edge(1, 2, 10)
g.add_edge(1, 3, 15)
g.add_edge(2, 3, 11)
g.add_edge(2, 5, 2)
g.add_edge(3, 4, 6)
g.add_edge(4, 5, 9)

# Find shortest paths from vertex 0
start_vertex = 0
distances = g.dijkstra(start_vertex)

# Print distances
print(f"Shortest paths from vertex {start_vertex}:")
for vertex, distance in distances.items():
    print(f"Vertex {vertex}: Distance {distance}")

# Visualize the graph
g.visualize_graph(start_vertex, distances)
```

# 12) Implement 0/1 knapsack using dynamic programming.

```python
def knapsack(values, weights, capacity):
    n = len(values)  # Number of items
    # Create a 2D DP array to store maximum value for each capacity and item
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    # Build the DP table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                # Item can be included: take the maximum of including or excluding the item
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                # Item cannot be included: take the value without including the item
                dp[i][w] = dp[i - 1][w]

    # The maximum value will be in the bottom-right corner of the table
    return dp[n][capacity], dp

# Example usage
values = [60, 100, 120]  # Values of the items
weights = [10, 20, 30]   # Weights of the items
capacity = 50            # Maximum capacity of the knapsack

max_value, dp_table = knapsack(values, weights, capacity)
print(f"Maximum value in Knapsack = {max_value}")

# Print the DP table for reference
print("\nDP Table:")
for row in dp_table:
    print(row)
```
